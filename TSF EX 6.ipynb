```python
# prompt: create dataset for me of flight dates and number of passensgers
import numpy as np
import pandas as pd
dates = pd.date_range('2023-01-01', periods=100, freq='D')
passengers = np.random.randint(100, 200, size=100)
data = {'Date': dates, 'Passengers': passengers}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.set_index('Date',inplace=True)
```

```python
df
```

```python
# prompt: create graph of this values

df['Passengers'].plot(figsize=(12, 6))
```

```python
df.rolling(3).mean()
```

```python
# prompt: create graph to show case rolling mean

df.rolling(3).mean().plot(figsize=(12, 6))
```

```python
df.expanding().mean()
```

```python
# prompt: create graph to show case ex[panding mean

df.expanding().mean().plot(figsize=(12, 6))
```

**We Idetify the downward trend in this**

```python
df.ewm(span=2).mean()
```

```python
# prompt: create graph to show case ewm mean

df.ewm(span=2).mean().plot(figsize=(12, 6))
```

```python
# prompt: time series exponetial smoothing

df['Passengers'].ewm(alpha=0.5).mean().plot(figsize=(12, 6))
```
