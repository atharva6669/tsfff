```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
np.random.seed(0)
n = 100
time = np.arange(n)
linear_trend = time + np.random.normal(scale=5,size=n)
linear_series = pd.Series(linear_trend)
```

```python
nonlinear_trend = np.sin(np.linspace(0, 4*np.pi, n)) + np.random.normal(scale=0.5, size=n)
nonlinear_series = pd.Series(nonlinear_trend)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(linear_series)
plt.title('Linear Trend')

plt.subplot(2,1,2)
plt.plot(nonlinear_series)
plt.title('Nonlinear tREND')

plt.tight_layout()
plt.show()
```

## Is there any difference between stationary time series and linear time series?

## **Stationary Time Series:**
A stationary time series is one where the variance in data remains constant over different periods of time.
Key characteristics of a stationary time series:
No seasonal fluctuation: The pattern of data remains consistent over time.
Autocorrelational: The data points exhibit correlation with their past values.
Mean and variance are constant: The average and spread of data do not change.

## **Linear Time Series:**
Linear time series refers to a specific class of time series models that are applicable only to linear systems.
They assume linearity, which may not hold for many real-world phenomena.
They cannot capture the behavior of nonlinear systems.
Yule’s 1927 analysis of the sunspot cycle was a pioneering example of inferring a linear autoregressive model from measurements of an external observable 1.

**In summary, while linear time series models are powerful, they are limited to linear systems. Stationary time series, on the other hand, focus on consistent patterns and variance stability, regardless of linearity or nonlinearity.**

```python
X_linear = time.reshape(-1,1)
y_linear = linear_series.values.reshape(-1,1)

linear_model = LinearRegression()
linear_model.fit(X_linear, y_linear)

linear_prediction = linear_model.predict(X_linear)

plt.figure(figsize=(8,4))
plt.plot(linear_series, label='Original Data')
plt.plot(linear_prediction, label='Linear Trend', linestyle='--')
plt.title('Linear Trend - Model')
plt.legend()
plt.show()
```

```python
from scipy.optimize import curve_fit

def sinusoidal_function(x, a, b, c, d):
    return a * np.sin(b*x+c)+d

popt, _ = curve_fit(sinusoidal_function, time, nonlinear_series)

nonlinear_predictions = sinusoidal_function(time, *popt)

plt.figure(figsize=(8,4))
plt.plot(nonlinear_series, label='Orignal Data')
plt.plot(nonlinear_predictions, label='Fitted Sinusoidal Pattern', linestyle='--')
plt.title('Nonlinear Trend - Fitted Sinusoidal Pattern')
plt.legend()
plt.show()
```
