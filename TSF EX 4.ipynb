```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
date_range = pd.date_range(start='2022-01-01',end='2024-01-01',freq='D')
values = np.sin(np.arange(len(date_range)) * np.pi / 180) + np.random.normal(0,0.1,size=len(date_range))
date = pd.DataFrame({'Date': date_range,'Value': values})
```

```python
date
```

```python
plt.figure(figsize=(10,6))
plt.plot(date['Date'], date['Value'], label='Orignal Data')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title("Orignal Time Series Data")
plt.legend()
plt.show()
```

```python
monthly_date = date.resample('M', on='Date').mean()
plt.figure(figsize=(10,6))
plt.plot(monthly_date.index, monthly_date['Value'], label='Aggregated Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Aggregated Time Series Data')
plt.legend()
plt.show()
```

```python
smoothed_date = date['Value'].rolling(window=10).mean()
plt.figure(figsize=(10,6))
plt.plot(date['Date'], smoothed_date, label='smoothed Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Smoothed Time Series Data')
plt.legend()
plt.show()
```

```python
ploy_coefficients = np.polyfit(date.index, date['Value'], deg=2)
poly_values = np.polyval(ploy_coefficients, date.index)
plt.figure(figsize=(10,6))
plt.plot(date['Date'], poly_values, label='Ploynomial Fitting')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Polynomaial Fitting Time Series Data')
plt.legend()
plt.show()
```

```python
ploy_coefficients = np.polyfit(date.index, date['Value'], deg=3)
poly_values = np.polyval(ploy_coefficients, date.index)
plt.figure(figsize=(10,6))
plt.plot(date['Date'], poly_values, label='Ploynomial Fitting')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Polynomaial Fitting Time Series Data')
plt.legend()
plt.show()
```
